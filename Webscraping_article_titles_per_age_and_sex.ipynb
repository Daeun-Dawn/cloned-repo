```python
from selenium import webdriver
ch_driver=webdriver.Chrome('C:/Users/user/chromedriver.exe')
url='https://media.daum.net/ranking/age/?regDate=20190103'
ch_driver.get(url)
```

```python
# 20대, 30대, 40대, 50대 별 가져오기
apath = '//*[@id="mArticle"]/div[2]/div'
per_age=ch_driver.find_elements_by_xpath(apath)
len(per_age)  # 4개
```

```python
#20대 중 여성이 많이 본 기사제목 가져오기
twenties_female=per_age[0].find_element_by_class_name('rank_female')
```

```python
twenties_female.text
```

```python
article_titles=twenties_female.text.split('\n')  # get_article_title 함수의 age_sex
article_titles
```

```python
# 기사 제목 뒤, 신문사/방송사 이름 제거하기
# article_title를 20대 여성, 남성/ 30대 여성, 남성 식으로 구분해
# 딕셔너리 형태로 다시 저장하기{20대 여성 : [ㅁㅁㅁㅁㅁㅁㅁㅁㅁ, ㅁㅁㅁㅁㅁㅁㅁㅁ,ㅁㅁㅁㅁㅁㅁㅁㅁ,] }
titles = dict()
for title in article_titles[1:]:
    if article_titles[0] not in titles:
        titles[article_titles[0]] = []
    titles[article_titles[0]].append(title[:-4].strip(" "))
    
titles            
    
```

<How to append multiple values for one key in dictionary>

def add_or_append(dictionary, key, value):

    if key not in dictionary:
    
        dictionary[key] = []
        
    dictionary[key].append(value)

```python
# 아래 코드는 리스트에 저장한 형식
article_title=[]
for title in article_titles[1:]:
    title= title[:-5]
    article_title.append(title)
article_title
```

```python
# make these to function
def get_article_title():
    apath = '//*[@id="mArticle"]/div[2]/div'
    per_age=ch_driver.find_elements_by_xpath(apath)
    female_title =dict()
    male_title=dict()
    for age in per_age:  # per age : 20대, 30대, 40대, 50대 
        age_sex= age.find_element_by_class_name('rank_female')
        article_titles=age_sex.text.split('\n')
        for title in article_titles[1:]:
            if article_titles[0] not in female_title:
                female_title[article_titles[0]] = []
            female_title[article_titles[0]].append(title[:-4].strip(" "))
            
        
        age_sex= age.find_element_by_class_name('rank_male')
        article_titles=age_sex.text.split('\n')
        for title in article_titles[1:]:
            if article_titles[0] not in male_title:
                male_title[article_titles[0]] = []
            male_title[article_titles[0]].append(title[:-4].strip(" "))
        
        
    return female_title, male_title

get_article_title()
```

```python
# 다음 페이지로 이동하며 총 20페이지 긁어모아오기
#20190103부터 20190122 까지 연령, 성별 기사 제목 긁기
from time import sleep

result_list=[]
def get_all_titles(date): # date == 20190103
    
    for specific_date in range(date, date+20):
        front_url = 'https://media.daum.net/ranking/age/?regDate='
        back_url = str(specific_date)
        ch_driver.get(front_url + back_url)

        sleep(1)

        result_list.append(get_article_title())

    return result_list

get_all_titles(20190103)
```
